## Limppieza de datos

```python
# Renombrando las columnas

dataFrame_original.rename(columns={'LIMIT_BAL':'CREDITO','SEX':'GENERO','EDUCATION':'EDUCACION',
                       'MARRIAGE':'ESTADO_CIVIL', 'AGE': 'EDAD', 'PAY_0': 'PAGO1',
                       'PAY_2': 'PAGO2','PAY_3': 'PAGO3','PAY_4': 'PAGO4',
                      'PAY_5': 'PAGO5','PAY_6': 'PAGO6', 'BILL_AMT1': 'SALDO1', 'BILL_AMT2': 'SALDO2',
                       'BILL_AMT3': 'SALDO3', 'BILL_AMT4': 'SALDO4', 'BILL_AMT5': 'SALDO5', 'BILL_AMT6': 'SALDO6',
                       'PAY_AMT1': 'PAGADO1','PAY_AMT2': 'PAGADO2', 'PAY_AMT3': 'PAGADO3','PAY_AMT4': 'PAGADO4',
                      'PAY_AMT5': 'PAGADO5','PAY_AMT6': 'PAGADO6', 'default payment next month': 'COMPORTAMIENTO'}, inplace=True)

# Eliminamos la columna "ID"
dataFrame_original = dataFrame_original.drop('ID', axis=1)
dataFrame = dataFrame_original
```

```python
# Mostrando los nuevos  nombres de las  columnas
dataFrame.columns
```

```python
# Mostrando tipos de datos del frame de datos
dataFrame.info()
```

```python
# Buscando y mostrando datos nulos o invalidos por COLUMNA
dataFrame.isnull().sum()
```

```python
# Buscando y mostrando TOTAL de datos nulos o invalidos

print("Total NaN datos en el Dataframe" , dataFrame.isnull().sum().sum(), sep='\n')
```

```python
# Convertimos genero a categoria
dataFrame['GENERO'] = dataFrame_original['GENERO'].map({1:'Hombre', 2:'Mujer'})
dataFrame['GENERO'] = dataFrame_original['GENERO'].astype('category')
# Convertimos educacion a categoria
dataFrame['EDUCACION'] = dataFrame_original['EDUCACION'].map({1:'Profecional', 2:'Universitario', 3:'Secundaria', 0:'Otro', 4:'Otro', 5:'Otro', 6:'Otro'})
dataFrame['EDUCACION'] = dataFrame_original['EDUCACION'].astype('category')
# Convertimos estadocivil a categoria
dataFrame['ESTADO_CIVIL'] = dataFrame_original['ESTADO_CIVIL'].map({1:'Casado', 2:'Soltero', 3:'Divorciado', 2:'Otro'})
dataFrame['ESTADO_CIVIL'] = dataFrame_original['ESTADO_CIVIL'].astype('category')
# Convertimos comportamiento a categoria
dataFrame['COMPORTAMIENTO'] = dataFrame_original['COMPORTAMIENTO'].map({1:'Moroso', 0:'Normal'})
dataFrame['COMPORTAMIENTO'] = dataFrame_original['COMPORTAMIENTO'].astype('category')
```

```python
#Funcion para clasificar las edades de lso clinetes

def clasificar_PErsona_edad(cliente):
    edad = cliente
    
    if edad < 28:
        return 'JOVEN'
    elif edad < 55:
        return 'ADULTO'
    else: return 'ADULTO_MAYOR'
    
```

```python
#Clasificamos la edad con la funcion que creamos anteriormente
dataFrame['TIPO_PERSONA'] = dataFrame['EDAD'].apply(clasificar_PErsona_edad)
```

```python
#Tiramos un resumen de los datos clasificados
dataFrame.groupby('TIPO_PERSONA')['TIPO_PERSONA'].count()
```

```python
#Grafico Cleintes agrupados por genero

def roundup(x): return 100 + int(math.ceil(x / 100.0)) * 100

fg = sns.catplot('TIPO_PERSONA', data=dataFrame, kind='count', aspect=1.75)
fg.set_xlabels('GENERO POR CLIENTES')
#fg.fig.suptitle('Clientes por Genero')

ax = plt.gca()
y_max = dataFrame['TIPO_PERSONA'].value_counts().max() 
ax.set_ylim([0, roundup(y_max)])

for p in ax.patches:
    ax.text(p.get_x() + p.get_width()/2., p.get_height(), '%d' % int(p.get_height()), 
            fontsize=12, color='red', ha='center', va='bottom')
```
